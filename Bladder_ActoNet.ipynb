```python
import actonet
import biolqm
```

```python
lqm = biolqm.load("http://ginsim.org/sites/default/files/Bladder_Model.zginml")
```

```python
inputs = {
    "GrowthInhibitors":1,
    "EGFR_stimulus":1,
    "FGFR3_stimulus":1
}
target = {"Apoptosis_b1": 1, "RB1": 1}
```

```python
an = actonet.ActoNet(lqm, inputs)
```

```python
s = an.reprogramming_fixpoints(target, maxsize=2)
s
```

```python
s.as_graph()
```
